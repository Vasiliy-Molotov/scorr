4.Предложить метрику (рассчитываемую с использованием предоставленных данных), с помощью которой можно было бы отранжировать друзей для выбранного человека по степени близости/силы дружбы с ним.
    
    (i,j - id двух выбранных друзей)
    Базовое предположение - чем больше у i и j общего, тем они более близкие друзья.
    В соответствии с этим создадим метрику на основании следующих индексов:
    
    index_friends = ((#общих друзей)/(#друзей i))*((#общих друзей)/(#друзей j)).
    Где (#общих друзей)/(#друзей i) отражает социальную значимость данной группы общих друзей для i. Социальная значимость имеет в числителе (#общих друзей), чтобы отражать прямую зависимость между общими друзьями и близостью дружбы, а в знаменателе (#друзей), чтобы учесть долю, которые эти друзья занимают в жизни человека. index_friends = 1, когда все друзья совпадают, и 0, когда их нет.
    
    index_age
    index age = - abs(age_i - age_j)
    Чем более близкий возраст - тем ближе социальная группа, и тем больше вероятность иметь больше общего и быть ближе.
    Если по age_i либо по  age_j отсутствуют данные, то вместо них можно брать оценку их возраста на основе возрастов их друзей.
    
    index_town_1 = (1, если town_i = town_j; 0 в другом случае)
    Если друзья из одного города - вероятность близкой дружбы больше.
    Если доступна информация по географическому положению городов, то вместо бинарного параметра можно выбрать: index_town_1 = -(расстояние между городами)
    Город, как и возраст, можно предсказывать, если отсутствует оригинальая информация.
    
    index_town_2 = (суммировать delta_k по всем городам k, в которых есть друзья i или j),
                    delta_k = ((#друзей i в k)/(#друзей i))*((#друзей j в k)/(#друзей j))
    если  i и j имеют схожую географию социальных взаимодействий - вероятность близкой дружбы выше. index_town_2 = 1, если все друзья i и j из одного города. index_town_2 = 0, если у i и j нет общих друзей ни в одном городе.
    
    Метрику построим следующим образом. Для каждого индекса отсортируем в порядке убывания его значений всех друзей заданного пользователя i. Таким образом для каждого его друга j получим его позиции в каждом списке: rank_friends(i,j), rank_index_age(i,j), ...
    Тогда метрикой близости дружбы назовем сумму порядковых позиций по всем спискам
    
    Friendship_power(i,j)  = rank_friends(i,j) + rank_index_age(i,j) + ...
    Чем меньше значение Friendshop_power, тем ближе связь между i и j.
    
    Плюс такой метрики:
    + она учитывает особенности заданной пары друзей по всем известным параметрам - количеству общих друзей, городу и возрасту.
    + в нее можно добавлять и убирать дополнительные переменные по указанной выше модели с сохранением существующей логики. Поэтому для человека, по которому, например, отсутствует информации по возрасту (и если не браться его предсказывать), или она имеется в наличии, но похоже на фальшивый возраст, то рейтинг можно построить на основании других оставшихся параметров.
    + хорошо предсказывает дружбу в социальных группах вида "школа" - общий возраст, город, общие друзья, "институт" - как школа, может отличаться город.
    
    
    Минусы:
    - значимость всех индексов признается одинаковой. Решение можно искать в области перехода к метрике вида линейной комбинации по каждому индексу, где основная задача - определить обоснованные весов.
    - Плохо предсказывает социальные связи вида - "родственник" и "отношения", где по всем параметрам могут быть существенные отличия, что может быть улучшено при учете дополнительных параметров - например, по индексу "совпадающая фамилия" и "семейный статус".
    
    
    Точность может быть повышена за счет дополнительных параметров, доступных в социальных сетях.
    - пол
    - место учебы/работы
    - расстояния между городами по геолокации
    - семейный статус
    - посты
    - группы
    - документы/музыка/видео
    
    Например информацию по состоянию в группах можно учитывать на основе индекса, предложенного в
    http://www.hpl.hp.com/research/idl/papers/web10/fnn2.pdf
    
    index_group(i,j) = (сумма delta(i,j) по всем группам, в которых состоит i и хотя бы один его друг)
                        delta(i,j) = 1 / log(frequency(group(i,j))) 
                        
                        frequency(group(i,j)) = 2, если в этой групе из друзей i состоит только j
                        frequency(group(i,j)) = 3, если в этой групе из друзей i состоит j и еще 1 друг i
                        frequency(group(i,j)) = 4, если в этой групе из друзей i состоит j и еще 2 друга i

```python
# initializing sparse matrix representation of an adjacency matrix S(i,j)
# i,j = id-s
# i x j = 1 if friends
S_size = 12000000  # this size is chosen to cover all id-s from input data
S = dok_matrix((S_size, S_size), dtype=np.int8)

# methods to save sparse matrices to disc to operate within memory limits

def save_sparse_csr(filename,array):
    np.savez(filename,data = array.data ,indices=array.indices,
             indptr =array.indptr, shape=array.shape )
    
def load_sparse_csr(filename):
    loader = np.load(filename)
    return csr_matrix((  loader['data'], loader['indices'], loader['indptr']),
                         shape = loader['shape'])
```

```python
#getting file names for vertices data
os.chdir("C:\Users\Vasiliy\Desktop\SCorr\DS_task\\adjacency_list")
v_file = []  # list for file names
for file in glob.glob("part*"):
    v_file.append(file)
```

```python
import math
#creating indices for KFold CV
def KFold(n,k,shuffle):  # for improvement - shuffling with the same portion of each class in each fold
    # n - length
    # k - number of folds
    a = np.array(range(n))
    l = math.floor(n/k)
    if l < 1:
        print('KFold: set is too small')
        a[:] = 1
    elif l == 1:
        a[:] = 1
    if l > 1:
        p = n - k*l
        i1 = int(n -(p)*(l+1))
        for i in range(0,i1-1):
            a[i] = math.floor(i/l) + 1
        for i in range(i1-1, n):
            a[i] = math.floor((i-i1)/(l+1)) + (k-p) + 1
    if shuffle == 1:
        a = np.random.permutation(a[:])
    return a
```

```python
from sklearn.metrics import mean_squared_error
def accuracy(x1,x2):
#     acc=0
#     for i in range (0,len(x1)):
#         if x1[i]==x2[i]: #not optimal performance - could be improved
#             acc+=1
    return mean_squared_error(x1,x2)
```

```python
def CV(model,X,y,k,text ='',baseline='',savename='CV.png',acc_array=list(),param_array=list()):
    graph_KFold = np.arange(k).astype(float)
    # X - train
    # y - fit
    # k - number of folds
    fic = 'KFold' #Fold_index_column 
    # I - array with neccesary indices for k-Fold CV
    I=KFold(X.shape[0],k,1) #link to KFold
    X_=X.copy()
    y_=y.copy()
    X_[fic] = I
    y_[fic] = I
    for i in range(1,k+1):
        X_train = X_[X_[fic] != i].drop(fic,1)
        y_train = y_[y_[fic] != i].drop(fic,1)
        X_test= X_[X_[fic] == i].drop(fic,1)
        y_test = y_[y_[fic] == i].drop(fic,1)    
        model.fit(X_train,y_train['year'].values) #insert target column name here
        predict = model.predict(X_test)
        graph_KFold[i-1] = accuracy(predict,list(y_test.ix[0:,0]))
        # print text,k,'fold CV, fold№ %d , acc=%f' % (i,graph_KFold[i-1])
    CV_plot(graph_KFold,k,baseline=baseline,savename=savename)
    print text,k,'fold CV, average acc = %f' % (graph_KFold.mean())
    acc_array.append(graph_KFold.mean())
    param_array.append(text)
    return acc_array, param_array
```

```python
def CV_plot(graph_KFold,k,savename='CV.png',baseline=''):
    # fig = plt.figure()
    ax = plt.subplot()                       #ax = plt.subplot(3,2,subplot_n)
    ax.bar(range(k), graph_KFold, width = 0.8)
    ax.axhline(graph_KFold.mean(), color='blue', linewidth=2) #ax.axhline(RMSD_KFold.mean(), color='blue', linewidth=2)
    if baseline <>'':
        ax.axhline(baseline, color='red', linewidth=2)
        ax.annotate("{0:.2f}".format(baseline),xy = (0,baseline),xytext = (0,5),textcoords='offset points')
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_title(savename)
    ax.annotate("{0:.2f}".format(graph_KFold.mean()),xy = (0,graph_KFold.mean()),xytext = (0,5),textcoords='offset points')
    for X, Y, Z in zip(np.arange(k),graph_KFold, ["%.2f" % v for v in graph_KFold.astype(float)]):
        ax.annotate(Z, xy = (X, Y),ha='left',xytext=(0,5),textcoords='offset points')
    os.chdir("C:\Users\Vasiliy\Desktop\SCorr\DS_task\\Pic\\CV")
    pylab.savefig(savename, bbox_inches='tight')
    plt.close()
    #return # ax #fig

    #float("{0:.2f}".format(x))
```
